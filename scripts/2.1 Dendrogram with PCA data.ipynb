## Dendrogram 
### Single Method

```python
distance_single = linkage(dfPCA.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9']],method="single")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method",fontsize=18)
plt.show()
```

### Complete Method

```python
distance_single = linkage(dfPCA.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9']],method="complete")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Complete Method",fontsize=18)
plt.show()
```

### Average Method

```python
distance_single = linkage(dfPCA.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9']],method="average")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Average Method",fontsize=18)
plt.show()
```

```python
# Importing pleasant weather dataset
pleasant_weather = pd.read_csv(os.path.join(path,'Dataset-Answers-Weather_Prediction_Pleasant_Weather.csv'))
pleasant_weather.head()
```

```python
# Reducing pleasant weather data to year 2020 only 
pw_2020 = pleasant_weather[pleasant_weather['DATE'].astype(str).str.contains('2020')]
pw_2020
```

```python
df_merged = pd.concat([dfPCA, pw_2020.reset_index(drop=True)], axis=1)
```

```python
df_merged
```

```python
from scipy.cluster.hierarchy import fcluster

# Use your hierarchical clustering linkage, e.g. distance_single or dist_sin
df_merged['clust2'] = fcluster(distance_single, 2, criterion='maxclust')
df_merged['clust3'] = fcluster(distance_single, 3, criterion='maxclust')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,6))
plt.suptitle("Hierarchical Clustering vs Pleasant Weather Labels", fontsize=18)

# Cluster K=2 plot
plt.subplot(1,3,1)
plt.title("K = 2 clusters", fontsize=14)
sns.scatterplot(data=df_merged, x='PCA1', y='PCA2', hue='clust2', palette='Set1')

# Actual pleasant weather labels plot (for one city, e.g., BASEL)
plt.subplot(1,3,3)
plt.title("BASEL Pleasant Weather", fontsize=14)
sns.scatterplot(data=df_merged, x='PCA1', y='PCA2', hue='BASEL_pleasant_weather', palette='coolwarm')
```

```python
plt.figure(figsize=(20, 5))
plt.title("Cluster vs Pleasant Weather for Basel (K = 2)", fontsize=14)
sns.swarmplot(
    x="BASEL_pleasant_weather",  # actual label (x-axis)
    y="clust2",                  # predicted cluster (y-axis)
    data=df_merged,
    hue="BASEL_pleasant_weather",  # optional color distinction
    dodge=True
)
```
