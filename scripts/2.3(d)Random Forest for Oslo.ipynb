```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn import datasets  
from sklearn.ensemble import RandomForestClassifier
from numpy import argmax
from sklearn.model_selection import train_test_split
from sklearn import metrics  
from sklearn.tree import plot_tree
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
# define path
path = r'/Users/tanu/Desktop/Data Analytics/ML with python/Data'
```

```python
# Impor6ting the clean weather data

df = pd.read_csv(os.path.join(path,'weather_cleaned.csv'))
df
```

```python
# Import the 'Pleasant weather' dataset

pw = pd.read_csv(os.path.join(path, 'Dataset-Answers-Weather_Prediction_Pleasant_Weather.csv'))
pw
```

## Data Preparation

```python
# Create a list of the columns containing "OSLO" in their names
oslo_list = list(x for x in df.columns if x.find('OSLO') >=0)
oslo_list
```

```python
# Create a Dataframe with these columns

df_oslo = df[oslo_list]
df_oslo
```

```python
# Reduce the pleasant weather data to Oslo answers only

pw_oslo = pw['OSLO_pleasant_weather']
pw_oslo
```

## Reshaping for Model

```python
X = np.array(df_oslo)
y = np.array(pw_oslo)
print (X.shape, y.shape)
```

## Data Split

```python
# Split data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## Random Forest Model

```python
# Create a RF classifier
clf = RandomForestClassifier(n_estimators = 100)#, max_depth=5)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
```

```python
# Perform predictions on the test dataset
y_pred = clf.predict(X_test)
  
# Use metrics module for accuracy calculation
print("Model Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# Class-names = {0:'Unpleasant Weather', 1:'Pleasant Weather'}

fig = plt.figure(figsize=(80,40))
plot_tree(clf.estimators_[6], fontsize = 20, filled=True);
```

```python
fig.savefig(os.path.join(path, 'Random_Forest_Oslo.png'),bbox_inches='tight')
```

## Feature Importance

```python
# Retrieve feature importances from the trained model

newarray = clf.feature_importances_
print(clf.feature_importances_.shape)
newarray
```

```python
# Create a list of weather features

wx_list = [feature.replace('oslo_', '') for feature in oslo_list]
wx_list
```

```python
important = pd.Series(newarray, index = wx_list)
important
```

```python
# Plot the results

%matplotlib inline

plt.style.use('fivethirtyeight')
# list of x locations for plotting
x_values = list(range(len(newarray)))
print(wx_list)

plt.bar(x_values, newarray, orientation = 'vertical')
plt.xticks(x_values, wx_list, rotation='vertical')
plt.ylabel('Importance'); plt.xlabel('Feature'); plt.title('Features Importances for Oslo');

plt.savefig(os.path.join(path, 'Feature_Importances_Oslo.png'), bbox_inches='tight')
plt.show()
```
